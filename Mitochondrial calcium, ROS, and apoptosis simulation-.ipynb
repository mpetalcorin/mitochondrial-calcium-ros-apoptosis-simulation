# Mitochondrial calcium, ROS, and apoptosis simulation
This workflow simulates biologically plausible single-cell data
for mitochondrial calcium signaling, ROS dynamics, mitochondrial
membrane potential, ATP, cytochrome c release, caspase activation,
and apoptosis probability under different experimental conditions.

## Important note:
- The dataset is synthetic.
- Values are normalized and anchored to qualitative and semi-quantitative trends reported in peer-reviewed PubMed-indexed literature.
- This is intended for computational exploration and proof-of-concept analysis, not as a replacement for real experimental measurements.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    roc_auc_score, accuracy_score, confusion_matrix, ConfusionMatrixDisplay,
    RocCurveDisplay
)
from sklearn.model_selection import train_test_split
from sklearn.inspection import permutation_importance

from scipy.special import expit
from scipy import stats


# Reproducibility

np.random.seed(42)


# Utility functions

def clipped_normal(mean, sd, n, low=None, high=None):
    x = np.random.normal(mean, sd, n)
    if low is not None:
        x = np.maximum(x, low)
    if high is not None:
        x = np.minimum(x, high)
    return x

def zscore_series(x):
    return (x - np.mean(x)) / np.std(x)

def minmax_0_1(x):
    x = np.asarray(x)
    return (x - x.min()) / (x.max() - x.min() + 1e-9)


# Biological design assumptions
# Conditions reflect literature-consistent scenarios:
# 1. Control
# 2. Mild stress
# 3. Calcium overload
# 4. ROS stress
# 5. Combined calcium + ROS stress
# 6. Antioxidant rescue
# 7. MCU inhibition / reduced mitochondrial calcium uptake

# Variables are in normalized units:
# - Cytosolic_Ca: relative cytosolic calcium signaling
# - Mito_Ca: relative mitochondrial calcium load
# - ROS: relative mitochondrial oxidative burden
# - DeltaPsi: relative membrane potential (1 = healthier, lower = depolarized)
# - ATP: relative ATP production
# - mPTP: permeability transition propensity
# - CytC_release: cytochrome c release score
# - Caspase3: caspase activation score
# - ApoptosisProb: probability from 0 to 1
# - Apoptotic: binary outcome

n_per_group = 500

conditions = {
    "Control": {
        "cyto_ca_mean": 0.35,
        "cyto_ca_sd": 0.07,
        "mito_ca_shift": 0.00,
        "ros_shift": 0.00,
        "delta_psi_shift": 0.00,
        "antioxidant_strength": 0.00,
        "mcu_scale": 1.00,
    },
    "Mild_Stress": {
        "cyto_ca_mean": 0.48,
        "cyto_ca_sd": 0.08,
        "mito_ca_shift": 0.08,
        "ros_shift": 0.10,
        "delta_psi_shift": -0.05,
        "antioxidant_strength": 0.00,
        "mcu_scale": 1.00,
    },
    "Ca_Overload": {
        "cyto_ca_mean": 0.72,
        "cyto_ca_sd": 0.10,
        "mito_ca_shift": 0.22,
        "ros_shift": 0.12,
        "delta_psi_shift": -0.12,
        "antioxidant_strength": 0.00,
        "mcu_scale": 1.12,
    },
    "ROS_Stress": {
        "cyto_ca_mean": 0.42,
        "cyto_ca_sd": 0.08,
        "mito_ca_shift": 0.06,
        "ros_shift": 0.28,
        "delta_psi_shift": -0.15,
        "antioxidant_strength": 0.00,
        "mcu_scale": 1.00,
    },
    "Combined_Ca_ROS": {
        "cyto_ca_mean": 0.78,
        "cyto_ca_sd": 0.10,
        "mito_ca_shift": 0.30,
        "ros_shift": 0.34,
        "delta_psi_shift": -0.22,
        "antioxidant_strength": 0.00,
        "mcu_scale": 1.15,
    },
    "Antioxidant_Rescue": {
        "cyto_ca_mean": 0.66,
        "cyto_ca_sd": 0.09,
        "mito_ca_shift": 0.18,
        "ros_shift": 0.06,
        "delta_psi_shift": -0.05,
        "antioxidant_strength": 0.18,
        "mcu_scale": 1.00,
    },
    "MCU_Inhibition": {
        "cyto_ca_mean": 0.70,
        "cyto_ca_sd": 0.10,
        "mito_ca_shift": -0.10,
        "ros_shift": 0.02,
        "delta_psi_shift": -0.03,
        "antioxidant_strength": 0.00,
        "mcu_scale": 0.58,
    },
}


# Simulation engine

rows = []

for condition, params in conditions.items():
    n = n_per_group

    # Cytosolic calcium signal
    cyt_ca = clipped_normal(
        params["cyto_ca_mean"], params["cyto_ca_sd"], n, low=0.05, high=1.30
    )

    # Mitochondrial calcium is driven by cytosolic calcium and MCU-dependent uptake
    mito_ca = (
        0.35 * cyt_ca * params["mcu_scale"]
        + params["mito_ca_shift"]
        + np.random.normal(0, 0.06, n)
    )
    mito_ca = np.clip(mito_ca, 0.02, 1.50)

    # ROS rises with mito calcium and condition-specific oxidative load
    ros = (
        0.28
        + 0.45 * mito_ca
        + params["ros_shift"]
        - params["antioxidant_strength"]
        + np.random.normal(0, 0.07, n)
    )
    ros = np.clip(ros, 0.02, 1.60)

    # Membrane potential declines with excessive mito calcium and ROS
    delta_psi = (
        1.00
        - 0.16 * mito_ca
        - 0.22 * ros
        + params["delta_psi_shift"]
        + np.random.normal(0, 0.05, n)
    )
    delta_psi = np.clip(delta_psi, 0.05, 1.20)

    # ATP rises with moderate mitochondrial calcium but falls with ROS and depolarization
    # This creates a biologically intuitive "sweet spot" for calcium.
    atp = (
        0.55
        + 0.28 * mito_ca
        - 0.24 * (mito_ca ** 2)
        - 0.30 * ros
        + 0.40 * delta_psi
        + np.random.normal(0, 0.05, n)
    )
    atp = np.clip(atp, 0.02, 1.30)

    # mPTP propensity rises with mito calcium overload, ROS, and loss of membrane potential
    mptp = (
        0.15
        + 0.55 * mito_ca
        + 0.45 * ros
        + 0.28 * (1 - delta_psi)
        + 0.20 * mito_ca * ros
        + np.random.normal(0, 0.06, n)
    )
    mptp = np.clip(mptp, 0.00, 2.00)

    # Cytochrome c release driven by mPTP and oxidative stress
    cytc_release = (
        -0.20
        + 0.80 * mptp
        + 0.20 * ros
        + np.random.normal(0, 0.07, n)
    )
    cytc_release = np.clip(cytc_release, 0.00, 2.00)

    # Caspase 3 activation depends on cytochrome c release and bioenergetic collapse
    caspase3 = (
        -0.10
        + 0.75 * cytc_release
        + 0.25 * (1 - atp)
        + np.random.normal(0, 0.06, n)
    )
    caspase3 = np.clip(caspase3, 0.00, 2.00)

    # Apoptosis probability from integrated stress signals
    logits = (
        -4.2
        + 2.2 * mito_ca
        + 2.0 * ros
        + 1.8 * (1 - delta_psi)
        + 1.6 * cytc_release
        + 1.3 * caspase3
        - 1.4 * atp
    )
    apoptosis_prob = expit(logits)
    apoptotic = np.random.binomial(1, apoptosis_prob, n)

    # Add some cell metadata
    cell_size = clipped_normal(1.00, 0.12, n, low=0.60, high=1.50)
    mitochondrial_mass = clipped_normal(1.00 + 0.10 * mito_ca.mean(), 0.12, n, low=0.55, high=1.60)

    for i in range(n):
        rows.append({
            "Condition": condition,
            "Cell_ID": f"{condition}_{i+1:03d}",
            "Cell_Size": cell_size[i],
            "Mito_Mass": mitochondrial_mass[i],
            "Cytosolic_Ca": cyt_ca[i],
            "Mito_Ca": mito_ca[i],
            "ROS": ros[i],
            "DeltaPsi": delta_psi[i],
            "ATP": atp[i],
            "mPTP": mptp[i],
            "CytC_Release": cytc_release[i],
            "Caspase3": caspase3[i],
            "ApoptosisProb": apoptosis_prob[i],
            "Apoptotic": apoptotic[i],
        })

df = pd.DataFrame(rows)

print("Dataset shape:", df.shape)
display(df.head())


# Summary statistics

summary = df.groupby("Condition").agg(
    n=("Cell_ID", "count"),
    Cytosolic_Ca_mean=("Cytosolic_Ca", "mean"),
    Mito_Ca_mean=("Mito_Ca", "mean"),
    ROS_mean=("ROS", "mean"),
    DeltaPsi_mean=("DeltaPsi", "mean"),
    ATP_mean=("ATP", "mean"),
    mPTP_mean=("mPTP", "mean"),
    CytC_Release_mean=("CytC_Release", "mean"),
    Caspase3_mean=("Caspase3", "mean"),
    ApoptosisProb_mean=("ApoptosisProb", "mean"),
    Apoptotic_rate=("Apoptotic", "mean"),
).round(3)

print("\nGroup summary:")
display(summary)


# Plot 1: Group means with error bars

features = ["Cytosolic_Ca", "Mito_Ca", "ROS", "DeltaPsi", "ATP", "mPTP", "CytC_Release", "Caspase3"]

fig, axes = plt.subplots(2, 4, figsize=(20, 9))
axes = axes.flatten()

for ax, feat in zip(axes, features):
    grp_mean = df.groupby("Condition")[feat].mean()
    grp_sem = df.groupby("Condition")[feat].sem()
    ax.bar(grp_mean.index, grp_mean.values, yerr=grp_sem.values, capsize=4)
    ax.set_title(feat)
    ax.tick_params(axis="x", rotation=45)

plt.tight_layout()
plt.show()


# Plot 2: Apoptosis rate by condition

apop_rates = df.groupby("Condition")["Apoptotic"].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
plt.bar(apop_rates.index, apop_rates.values)
plt.ylabel("Fraction apoptotic")
plt.title("Apoptotic fraction across conditions")
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.tight_layout()
plt.show()


# Plot 3: Correlation heatmap
# -----------------------------
corr_cols = ["Cytosolic_Ca", "Mito_Ca", "ROS", "DeltaPsi", "ATP", "mPTP", "CytC_Release", "Caspase3", "ApoptosisProb"]
corr = df[corr_cols].corr()

plt.figure(figsize=(9, 7))
plt.imshow(corr, aspect="auto")
plt.colorbar(label="Correlation")
plt.xticks(range(len(corr_cols)), corr_cols, rotation=45, ha="right")
plt.yticks(range(len(corr_cols)), corr_cols)
plt.title("Correlation matrix")
plt.tight_layout()
plt.show()

display(corr.round(3))

# -----------------------------
# Plot 4: Key biological scatter plots
# -----------------------------
sample_df = df.sample(1500, random_state=42)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

axes[0, 0].scatter(sample_df["Mito_Ca"], sample_df["ROS"], alpha=0.4)
axes[0, 0].set_xlabel("Mito_Ca")
axes[0, 0].set_ylabel("ROS")
axes[0, 0].set_title("Mitochondrial Ca vs ROS")

axes[0, 1].scatter(sample_df["ROS"], sample_df["DeltaPsi"], alpha=0.4)
axes[0, 1].set_xlabel("ROS")
axes[0, 1].set_ylabel("DeltaPsi")
axes[0, 1].set_title("ROS vs membrane potential")

axes[1, 0].scatter(sample_df["Mito_Ca"], sample_df["ATP"], alpha=0.4)
axes[1, 0].set_xlabel("Mito_Ca")
axes[1, 0].set_ylabel("ATP")
axes[1, 0].set_title("Mitochondrial Ca vs ATP")

axes[1, 1].scatter(sample_df["CytC_Release"], sample_df["Caspase3"], alpha=0.4)
axes[1, 1].set_xlabel("Cytochrome c release")
axes[1, 1].set_ylabel("Caspase3")
axes[1, 1].set_title("Cytochrome c release vs Caspase 3")

plt.tight_layout()
plt.show()

# -----------------------------
# Hypothesis tests across conditions
# -----------------------------
anova_results = []
for feat in ["Cytosolic_Ca", "Mito_Ca", "ROS", "DeltaPsi", "ATP", "mPTP", "CytC_Release", "Caspase3", "ApoptosisProb"]:
    groups = [grp[feat].values for _, grp in df.groupby("Condition")]
    f_stat, p_val = stats.f_oneway(*groups)
    anova_results.append({"Feature": feat, "F_stat": f_stat, "p_value": p_val})

anova_df = pd.DataFrame(anova_results).sort_values("p_value")
print("One-way ANOVA across conditions:")
display(anova_df)

# -----------------------------
# Pairwise comparison: Control vs Combined_Ca_ROS
# -----------------------------
pairwise_features = ["Mito_Ca", "ROS", "DeltaPsi", "ATP", "mPTP", "CytC_Release", "Caspase3", "ApoptosisProb"]

control = df[df["Condition"] == "Control"]
combined = df[df["Condition"] == "Combined_Ca_ROS"]

pairwise_stats = []
for feat in pairwise_features:
    t_stat, p_val = stats.ttest_ind(control[feat], combined[feat], equal_var=False)
    effect = (combined[feat].mean() - control[feat].mean()) / np.sqrt(
        (control[feat].var() + combined[feat].var()) / 2
    )
    pairwise_stats.append({
        "Feature": feat,
        "Control_mean": control[feat].mean(),
        "Combined_mean": combined[feat].mean(),
        "t_stat": t_stat,
        "p_value": p_val,
        "Cohens_d": effect,
    })

pairwise_df = pd.DataFrame(pairwise_stats).sort_values("p_value")
print("Control vs Combined_Ca_ROS comparison:")
display(pairwise_df.round(4))

# -----------------------------
# PCA on cellular states
# -----------------------------
pca_features = ["Cytosolic_Ca", "Mito_Ca", "ROS", "DeltaPsi", "ATP", "mPTP", "CytC_Release", "Caspase3"]
X = df[pca_features].values
X_scaled = StandardScaler().fit_transform(X)

pca = PCA(n_components=2)
pcs = pca.fit_transform(X_scaled)

pca_df = df[["Condition", "Apoptotic"]].copy()
pca_df["PC1"] = pcs[:, 0]
pca_df["PC2"] = pcs[:, 1]

plt.figure(figsize=(10, 7))
for cond in pca_df["Condition"].unique():
    subset = pca_df[pca_df["Condition"] == cond]
    plt.scatter(subset["PC1"], subset["PC2"], alpha=0.35, label=cond)
plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% var)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% var)")
plt.title("PCA of mitochondrial stress state")
plt.legend(bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()

loadings = pd.DataFrame(
    pca.components_.T,
    index=pca_features,
    columns=["PC1_loading", "PC2_loading"]
)
print("PCA loadings:")
display(loadings.round(3))

# -----------------------------
# Logistic regression: predict apoptosis
# -----------------------------
model_features = ["Cytosolic_Ca", "Mito_Ca", "ROS", "DeltaPsi", "ATP", "mPTP", "CytC_Release", "Caspase3"]
X = df[model_features]
y = df["Apoptotic"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

log_reg = LogisticRegression(max_iter=300)
log_reg.fit(X_train_s, y_train)

y_pred = log_reg.predict(X_test_s)
y_proba = log_reg.predict_proba(X_test_s)[:, 1]

print("Logistic regression performance")
print("AUC:", round(roc_auc_score(y_test, y_proba), 3))
print("Accuracy:", round(accuracy_score(y_test, y_pred), 3))

coef_df = pd.DataFrame({
    "Feature": model_features,
    "Coefficient": log_reg.coef_[0]
}).sort_values("Coefficient", ascending=False)

display(coef_df.round(3))

fig, ax = plt.subplots(figsize=(6, 5))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, ax=ax)
plt.title("Confusion matrix, logistic regression")
plt.show()

fig, ax = plt.subplots(figsize=(6, 5))
RocCurveDisplay.from_predictions(y_test, y_proba, ax=ax)
plt.title("ROC curve, logistic regression")
plt.show()

# -----------------------------
# Random forest: non-linear feature importance
# -----------------------------
rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=6,
    min_samples_leaf=8,
    random_state=42
)
rf.fit(X_train, y_train)
rf_proba = rf.predict_proba(X_test)[:, 1]
rf_pred = rf.predict(X_test)

print("Random forest performance")
print("AUC:", round(roc_auc_score(y_test, rf_proba), 3))
print("Accuracy:", round(accuracy_score(y_test, rf_pred), 3))

rf_importance = pd.DataFrame({
    "Feature": model_features,
    "Importance": rf.feature_importances_
}).sort_values("Importance", ascending=False)

display(rf_importance.round(3))

plt.figure(figsize=(8, 5))
plt.bar(rf_importance["Feature"], rf_importance["Importance"])
plt.xticks(rotation=45, ha="right")
plt.ylabel("Random forest importance")
plt.title("Feature importance for apoptosis classification")
plt.tight_layout()
plt.show()

perm = permutation_importance(rf, X_test, y_test, n_repeats=10, random_state=42)
perm_df = pd.DataFrame({
    "Feature": model_features,
    "PermutationImportance": perm.importances_mean
}).sort_values("PermutationImportance", ascending=False)

display(perm_df.round(3))

# -----------------------------
# Simple mediation-style analysis
# Does mito calcium affect apoptosis probability partly through ROS?
# -----------------------------
# Path A: Mito_Ca -> ROS
Xa = df[["Mito_Ca"]]
ya = df["ROS"]
reg_a = LinearRegression().fit(Xa, ya)
a_coef = reg_a.coef_[0]

# Path B and direct effect: Mito_Ca + ROS -> ApoptosisProb
Xb = df[["Mito_Ca", "ROS"]]
yb = df["ApoptosisProb"]
reg_b = LinearRegression().fit(Xb, yb)
b_ros = reg_b.coef_[1]
direct_mito = reg_b.coef_[0]
indirect_effect = a_coef * b_ros

mediation_df = pd.DataFrame({
    "Quantity": ["a: Mito_Ca -> ROS", "b: ROS -> ApoptosisProb | Mito_Ca", "Direct: Mito_Ca -> ApoptosisProb | ROS", "Indirect a*b"],
    "Estimate": [a_coef, b_ros, direct_mito, indirect_effect]
})
print("Simple mediation-style decomposition:")
display(mediation_df.round(4))

# -----------------------------
# Quadratic calcium-ATP relationship
# -----------------------------
# Demonstrates that moderate mitochondrial calcium can support ATP,
# while excessive calcium becomes detrimental.
quad_df = pd.DataFrame({
    "Mito_Ca": df["Mito_Ca"],
    "Mito_Ca_sq": df["Mito_Ca"] ** 2,
    "ATP": df["ATP"]
})

quad_model = LinearRegression().fit(quad_df[["Mito_Ca", "Mito_Ca_sq"]], quad_df["ATP"])
print("Quadratic ATP model coefficients:")
print("Intercept:", round(quad_model.intercept_, 4))
print("Mito_Ca:", round(quad_model.coef_[0], 4))
print("Mito_Ca_sq:", round(quad_model.coef_[1], 4))

x_curve = np.linspace(df["Mito_Ca"].min(), df["Mito_Ca"].max(), 200)
y_curve = (
    quad_model.intercept_
    + quad_model.coef_[0] * x_curve
    + quad_model.coef_[1] * (x_curve ** 2)
)

plt.figure(figsize=(8, 5))
plt.scatter(df.sample(1200, random_state=1)["Mito_Ca"], df.sample(1200, random_state=1)["ATP"], alpha=0.25)
plt.plot(x_curve, y_curve, linewidth=3)
plt.xlabel("Mito_Ca")
plt.ylabel("ATP")
plt.title("Biphasic relationship between mitochondrial calcium and ATP")
plt.tight_layout()
plt.show()

# -----------------------------
# Cell-state ranking
# Create a composite stress score
# -----------------------------
df["StressScore"] = (
    0.22 * zscore_series(df["Mito_Ca"])
    + 0.24 * zscore_series(df["ROS"])
    + 0.18 * zscore_series(1 - df["DeltaPsi"])
    + 0.18 * zscore_series(df["mPTP"])
    + 0.10 * zscore_series(df["CytC_Release"])
    + 0.08 * zscore_series(df["Caspase3"])
)

stress_summary = df.groupby("Condition").agg(
    StressScore_mean=("StressScore", "mean"),
    ApoptosisProb_mean=("ApoptosisProb", "mean"),
    Apoptotic_rate=("Apoptotic", "mean")
).sort_values("StressScore_mean", ascending=False)

print("Integrated stress ranking by condition:")
display(stress_summary.round(3))

plt.figure(figsize=(10, 5))
plt.bar(stress_summary.index, stress_summary["StressScore_mean"])
plt.xticks(rotation=45)
plt.ylabel("Mean integrated stress score")
plt.title("Condition ranking by integrated mitochondrial stress")
plt.tight_layout()
plt.show()

# -----------------------------
# Responder analysis
# Define "high-resilience" cells under stress
# -----------------------------
stress_conditions = df[df["Condition"].isin(["Mild_Stress", "Ca_Overload", "ROS_Stress", "Combined_Ca_ROS"])]
resilient = stress_conditions[
    (stress_conditions["ATP"] > stress_conditions["ATP"].quantile(0.75)) &
    (stress_conditions["ApoptosisProb"] < stress_conditions["ApoptosisProb"].quantile(0.25))
].copy()

vulnerable = stress_conditions[
    (stress_conditions["ATP"] < stress_conditions["ATP"].quantile(0.25)) &
    (stress_conditions["ApoptosisProb"] > stress_conditions["ApoptosisProb"].quantile(0.75))
].copy()

resilience_compare = pd.DataFrame({
    "Feature": ["Mito_Ca", "ROS", "DeltaPsi", "ATP", "mPTP", "CytC_Release", "Caspase3"],
    "Resilient_mean": [resilient[c].mean() for c in ["Mito_Ca", "ROS", "DeltaPsi", "ATP", "mPTP", "CytC_Release", "Caspase3"]],
    "Vulnerable_mean": [vulnerable[c].mean() for c in ["Mito_Ca", "ROS", "DeltaPsi", "ATP", "mPTP", "CytC_Release", "Caspase3"]],
})

print("Resilient vs vulnerable stressed cells:")
display(resilience_compare.round(3))

# -----------------------------
# Final export
# -----------------------------
summary.to_csv("mitochondrial_signal_summary.csv")
df.to_csv("simulated_mitochondrial_calcium_ros_apoptosis_dataset.csv", index=False)

print("Saved:")
print("- mitochondrial_signal_summary.csv")
print("- simulated_mitochondrial_calcium_ros_apoptosis_dataset.csv")
```
